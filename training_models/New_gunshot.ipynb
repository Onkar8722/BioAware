```python
import os
import librosa
import numpy as np

# REPLACE THIS with the actual path to ONE of your wav files
# Example: 'Sound/training/gunshot/shot001.wav'
TEST_FILE = "C:/Users/NSPatil/OneDrive/Desktop/SEM VI/Sound/Sound/Training/Gunshot/5B9FE452_3845e1bc9b-1037-41ed-8483-3212df99879015d3941d-e344-470d-9454-df7ed69b0b120bc76ac8-1ec5-420d-a3e1-f9083aca1f5f.WAV" 

# 1. Check if file exists
if not os.path.exists(TEST_FILE):
    print(f"ERROR: Could not find file at {TEST_FILE}")
    print("Please update the TEST_FILE variable with a real path from your folder.")
else:
    print(f"Found file: {TEST_FILE}")
    
    # 2. Try to load it with Librosa
    try:
        # Load audio
        audio, sample_rate = librosa.load(TEST_FILE, res_type='kaiser_fast')
        print(f"Audio loaded successfully! Sample Rate: {sample_rate}")
        
        # 3. Extract MFCC features (The "Fingerprint")
        mfccs = librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40)
        
        # 4. Take the average (mean)
        mfccs_scaled = np.mean(mfccs.T, axis=0)
        
        print(f"MFCC Feature Shape: {mfccs_scaled.shape}")
        print("Success! Your environment is ready for the next step.")
        print(f"First 5 features: {mfccs_scaled[:5]}")
        
    except Exception as e:
        print(f"Error processing audio: {e}")
```
